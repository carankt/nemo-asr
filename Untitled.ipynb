```python
import nemo
import nemo.collections.asr as nemo_asr
quartznet = nemo_asr.models.EncDecCTCModel.from_pretrained(model_name="QuartzNet15x5Base-En")
```

```python
file_names = []
with open("/dataset/colin/bad_file_names.txt", "r") as f:
    for lines in f:
        file_names.append("/dataset/colin/wavs_with_numbers/" + lines.split("|")[0])
```

```python
files = file_names
for fname, transcription in zip(files, quartznet.transcribe(paths2audio_files=files)):
    with open("/dataset/colin/file_with_numbers_as_text.txt", "+a") as f_:
        content = fname.split("/")[-1] +  "|"  + transcription + "\n"
        print(content)
        f_.write(content)
    print(f"Audio in {fname} was recognized as: {transcription}")
```
